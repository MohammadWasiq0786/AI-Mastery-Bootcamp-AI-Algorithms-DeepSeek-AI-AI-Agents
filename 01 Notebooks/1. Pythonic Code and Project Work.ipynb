### map(), filter(), reduce()

* **`map()`**
  * Applies a function to each item in an iterable.
* **`filter()`**
  * Filters items based on conditions.
* **`reduce()`**
  * Reduces an iterable to a single value.

```python
numbers= [1, 2, 3, 4]
sqaures= map(lambda x: x**2, numbers)
list(sqaures)
```

```python
numbers= [1, 2, 3, 4]
evenList= filter(lambda x: x%2 == 0, numbers)
list(evenList)
```

```python
from functools import reduce

numbers= [1, 2, 3, 4]
product= reduce(lambda x, y: x*y, numbers)
product
```

## Task Manager

```python
from random import choice
import os

# File to store tasks

FILE_NAME= "tasks.txt"

# Load tasks from file
def load_taks():
  tasks= {}
  if os.path.exists(FILE_NAME):
    with open(FILE_NAME, "r") as file:
      for line in file:
        task_id, title, status= line.strip(" | ")
        tasks[int(task_id)]= {'title': title, 'status': status}

  return tasks

# Save tasks to file

def save_tasks(tasks):
  with open(FILE_NAME, "w") as file:
    for task_id, task in tasks.items():

      file.write(f"{task_id} | {task['title']} | {task['status']} \n")

# Add a new task
def add_task(tasks):
  title= input("Enter task title: ")
  task_id= max(tasks.keys(), default=0) + 1
  tasks[task_id]= {'title': title, 'status': "Incomplete"}
  print(f"Task '{title}' added.")

# View all tasks
def view_tasks(tasks):
  if not tasks:
    print("No tasks available.")
  else:
    for task_id, task in tasks.items():
      print(f"[{task_id}] {task['title']} - {task['status']}")

# Maark task as complete
def mark_task_complete(tasks):
  task_id= int(input("Enter task ID to mark as complete: "))
  if task_id in tasks:
    tasks[task_id]['status']= 'Complete'
    print(f"Task '{tasks[task_id]['title']}' marked as complete.")
  else:
    print("Task ID not found")

# Maark task as complete
def delete_task(tasks):
  task_id= int(input("Enter task ID to mark as Delete: "))
  if task_id in tasks:
    delete_task= tasks.pop(task_id)
    print(f"Task '{delete_task['title']}' deleted.")
  else:
    print("Task ID not found")

# Main Menu
def main():
  tasks= load_taks()
  while True:
    print("\n Task Manager Manager Menu")
    print("1. Add Task")
    print("2. View Tasks")
    print("3. Mark Task as Complete")
    print("4. Delete Task")
    print("5. Exit")

    choice= input("Enter Your Choice: ")
    if choice == "1":
      add_task(tasks)
    elif choice == "2":
      view_tasks(tasks)
    elif choice == "3":
      mark_task_complete(tasks)
    elif choice == "4":
      delete_task(tasks)
    elif choice == "5":
      save_tasks(tasks)

      print("Goodbye")
      break

    else:
      print("Invalid Choice. Please try again")

if __name__ == "__main__":
  main()
```
